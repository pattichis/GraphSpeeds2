# #1. Understanding rates

a. A police car drives 236 km in 3 hours.
What is its average speed in kilometers per hour?

b. Mike drives his car 60 km in 55 minutes. What is its average speed in kilometers per minute?

````python

# Store the values in variables
distance = 1
time = 2

# Divide
rate = distance / time # Note the division!

# print the calculation
print("rate = distance / time")
print("distance  = ", distance) # Note how to print two things using ,
print("time = ", time)
print("rate = ", rate)
````

# #3 Assignment: Compute times from rates
